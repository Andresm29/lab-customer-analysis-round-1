```python
#Read the three files into python as dataframes
import pandas as pd
df1 = pd.read_csv("file1.csv")
df2 = pd.read_csv("file2.csv")
df3 = pd.read_csv("file3.csv")
```

```python
#Show the DataFrame's shape
df1.shape
```

```python
df2.shape
```

```python
df3.shape
```

```python
#Concatenate the three dataframes
data = pd.concat([df1, df2, df3])
data
```

```python
data.columns
```

```python
#Standardize header names
cols = [col_name.lower().replace(' ', '_') for col_name in data.columns]
cols
```

```python
data.columns = cols
data
```

```python
#Which columns are numerical?
#Which columns are categorical?

data.info()
```

```python
#Delete the column education and the number of open complaints from the dataframe.
data_new = data.drop(columns = ["education","number_of_open_complaints"])
data_new.columns
```

```python
#Correct the values in the column customer lifetime value. 
#They are given as a percent, so multiply them by 100 and change dtype to numerical type
data["customer_lifetime_value"] = pd.to_numeric(data["customer_lifetime_value"], errors='coerce')
```

```python
data["customer_lifetime_value"] = data["customer_lifetime_value"] * 100
```

```python
data["customer_lifetime_value"] = data["customer_lifetime_value"].astype(float)
```

```python
data
```

```python
#Check for duplicate rows in the data and remove if any.

duplicates = data[data.duplicated(keep=False)]
```

```python
if duplicates.shape[0] > 0:
    print("Duplicate rows:")
    print(duplicates)
else:
    print("No duplicate rows found.")
```

```python
#Filter out the data for customers who have an income of 0 or less.

data = data[data["income"] > 0]
data
```
